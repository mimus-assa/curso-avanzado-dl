```python
from tensorflow import *
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(128, 128,3)),
  tf.keras.layers.Dense(32),
  tf.keras.layers.Dense(32),
  tf.keras.layers.Dense(1, activation="sigmoid")
])
```

```python
tf.keras.utils.plot_model(model, show_shapes=True)
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.LSTM(32,input_shape=(32, 16, ), return_sequences=True),
  tf.keras.layers.LSTM(32, return_sequences=True),
  tf.keras.layers.LSTM(32, return_sequences=False),
  tf.keras.layers.Dense(1, activation="sigmoid")
    
])
```

```python
tf.keras.utils.plot_model(model, show_shapes=True)
```

```python


input_seq = tf.keras.layers.Input(shape=(32, 16, ))
x =  tf.keras.layers.LSTM(32, return_sequences=True)(input_seq)
x =  tf.keras.layers.LSTM(32, return_sequences=True)(x)
x =  tf.keras.layers.LSTM(32, return_sequences=False)(x)
out= tf.keras.layers.Dense(1, activation="sigmoid")(x)

model = tf.keras.Model(inputs=input_seq, outputs=out)

```

```python
tf.keras.utils.plot_model(model, show_shapes=True)
```

```python
x_1 = tf.keras.layers.Input(shape=(32, 16, ))
x_2 = tf.keras.layers.Input(shape=(32, 16, ))
Lstm1 =  tf.keras.layers.LSTM(32, return_sequences=True)(x_1)
Lstm1 =  tf.keras.layers.LSTM(32, return_sequences=True)(Lstm1)

Lstm2 =  tf.keras.layers.LSTM(32, return_sequences=True)(x_2)
Lstm2 =  tf.keras.layers.LSTM(32, return_sequences=True)(Lstm2)

concat_1_2 = tf.keras.layers.concatenate([Lstm1,Lstm2])
Lstm3 =  tf.keras.layers.LSTM(32, return_sequences=False)(concat_1_2)
                         
out= tf.keras.layers.Dense(1, activation="sigmoid")(Lstm3)

model = tf.keras.Model(inputs=[x_1, x_2], outputs=out)
```

```python
tf.keras.utils.plot_model(model, show_shapes=True)
```

```python
x_1 = tf.keras.layers.Input(shape=(32, 16, ))

Lstm1 =  tf.keras.layers.LSTM(32, return_sequences=True)(x_1)
Lstm1 =  tf.keras.layers.LSTM(32, return_sequences=False)(Lstm1)
                         
out1= tf.keras.layers.Dense(1, activation="sigmoid")(Lstm1)
out2= tf.keras.layers.Dense(10, activation="sigmoid")(Lstm1)

model = tf.keras.Model(inputs=x_1, outputs=[out1, out2])
```

```python
tf.keras.utils.plot_model(model, show_shapes=True)
```
